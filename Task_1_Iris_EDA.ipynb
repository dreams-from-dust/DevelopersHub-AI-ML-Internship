Libraries and Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set professional plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (9, 6)
plt.rcParams["font.size"] = 12

print("--- [Step 1] Loading Dataset ---")
df = sns.load_dataset('iris')

print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns\n")
print("--- [Step 2] First 5 Rows ---")
print(df.head())

print("\n--- [Step 3] Summary Statistics ---")
print(df.describe(include='all'))
```

Visualizations (Scatter, Histogram, Boxplot)

```python
# 1. Scatter Plot: Analyzing relationships between Petal Length and Petal Width
plt.figure()
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='deep', s=70, alpha=0.8)
plt.title('Petal Length vs. Petal Width by Species', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.legend(title='Species', frameon=True)
plt.tight_layout()
plt.show()

# 2. Histograms: Evaluating value distributions for Sepal Length
plt.figure()
sns.histplot(data=df, x='sepal_length', hue='species', kde=True, element='step', stat='density', common_norm=False, palette='muted')
plt.title('Distribution of Sepal Length Across Species (with KDE)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
plt.show()

# 3. Box Plots: Identifying outliers across Sepal Width
plt.figure()
sns.boxplot(data=df, x='species', y='sepal_width', palette='pastel', hue='species', legend=False)
plt.title('Sepal Width Box Plot: Outlier Identification', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.tight_layout()
plt.show()
```

Printed Insights

```python
print("INSIGHT (Scatter Plot): The scatter plot reveals distinct clustering. 'setosa' forms a completely isolated cluster with much smaller petals, while 'versicolor' and 'virginica' show a slight overlap, indicating a strong correlation between petal dimensions across species.\n")
print("INSIGHT (Histogram): The Kernel Density Estimate (KDE) lines show that sepal lengths follow rough normal distributions per species. 'virginica' consistently displays the longest sepals, while 'setosa' exhibits the shortest, more tightly grouped values.\n")
print("INSIGHT (Box Plot): The box plot allows us to spot data outliers easily. Notice the isolated data points above and below the whiskers for the 'setosa' species—these represent statistical outliers where the sepal width is unusually large or small compared to the rest of its group.\n")
```
